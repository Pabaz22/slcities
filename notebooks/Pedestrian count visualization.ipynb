## Nettoyage DF senseurs

```python
sensor_df = pd.read_csv('../raw_data/sensors_location.csv')
```

### Je vire les colonnes inutiles et je range un peu le df

```python
sensor_df = sensor_df.drop(columns=['User Type','Setup Date']).drop([4,8]).sort_values(by='Counter Locations').reset_index(drop=True)
```

### Je crée une colonne avec les coordonnées géospatiales en format geopandas

```python
sensor_gdf = gpd.GeoDataFrame(
    sensor_df, geometry=gpd.points_from_xy(sensor_df.Longitude, sensor_df.Latitude))
```

```python
sensor_gdf = sensor_gdf.reset_index().rename(columns={'index':'sensor_index'})
```

### Je génère les ID pour chacune de rues

```python
sensor_gdf['sensor_ID'] =  \
sensor_gdf['sensor_index'].apply(lambda x :'ID_' + '0' + str(sensor_gdf['sensor_index'][x]) if x <10\
                                 else 'ID_' + str(sensor_gdf['sensor_index'][x]))
```

```python
sensor_gdf = sensor_gdf.drop(columns='sensor_index')
```

```python
sensor_gdf
```

### Je récupère un dictionnaire pour renommer mes rues dans le DF de comptage

```python
dict_sensors_ID = {}
```

```python
for i in range(sensor_gdf.shape[0]):

    dict_sensors_ID[sensor_gdf['Counter Locations'][i]] = str(sensor_gdf['sensor_ID'][i])
```

```python
dict_sensors_ID
```

### On range un peu

```python
df_count = df_count.rename(columns=dict_sensors_ID)
```

```python
df_count = df_count.reindex(sorted(df_count.columns), axis=1)
```

```python
df_count = df_count.rename(columns= {'Date and Time' : 'date and time'})
```

```python
sensor_gdf = sensor_gdf.rename(columns={'Counter Locations':'counter locations', 'Latitude':'latitude','Longitude':'longitude'})
```
